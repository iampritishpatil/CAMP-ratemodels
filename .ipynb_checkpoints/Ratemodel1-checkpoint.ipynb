```python
%matplotlib inline
import math
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
global c1,c2,c3,c4,ke,ki,r,tau_e,tau_i,a,theta,P,Q
c1=12
c2=4
c3=13
c4=11
a=1
theta=1
ke=1
ki=1
re=1
ri=1
r=1
P=0
Q=0

def eqs(y,t):
    E,I=y
    dEdt =  -E + sigm(c1*E - c2*I + P)*(ke-re*E)
    dIdt =  -I + sigm(c3*E - c4*I + Q)*(ki-ri*I)
    dydt=[dEdt,dIdt]
    return dydt
```

```python
sigm = lambda x : 1 / (1 + math.exp(-1*a*(x-theta))) - 1 / (1 + math.exp(-1*a*(theta)))
```

```python
y0=[0,0]
t = np.linspace(0, 10, 1001)
sol = odeint(eqs, y0, t)
```

```python
import matplotlib.pyplot as plt
plt.plot(t, sol[:, 0], 'b', label='E')
plt.plot(t, sol[:, 1], 'g', label='I')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```

## Building upto a network model

We first start with a simple ratemodel of a neuron.

$ \tau \dot{w} = -w + i $ where $i$ is input

```python
def eqs(y,t):
    dydt=(-y + I(t))/tau()
    return dydt
tau = lambda : 10
I = lambda t : t*0.1
# I = lambda t : 0.5*(t>5 and t<7)
I = lambda t : np.sin(t/2.)*0.1


y0=[0.]
t = np.linspace(0, 20, 1001)
sol = odeint(eqs, y0, t)
plt.style.use('ggplot')
plt.plot(t, sol[:, 0], 'b')
plt.plot(t, I(t), 'r')
# plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```

Lets make a neuron with an autoapse, a synapse with itself

$ \tau \dot{y} = -y + Wy + i $

```python
def eqs(y,t):
    tau=20
    W=0.9
    dydt=(-y  + W * y + I(t))/tau
    return dydt

# def I(t):
#     return 0.1

def I(t):
    if t>500 and t<700:
        return 0.1
    return 0

# def I(t):
#     T=200
#     return 0.1*np.sin(2*np.pi*t/T)


y0=[0.00]
tmax=1.
t = np.linspace(0, tmax*1000, tmax*1000+1)
sol = odeint(eqs, y0, t , hmax=1)
plt.style.use('ggplot')
plt.plot(t, sol[:, 0], 'b',label="Output")
plt.plot(t, map(I,t), 'r',label="Input")
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```

## Mutually Inhibitory Pair

$ \tau \dot{x_1} = -\bar{x_1} - \beta x_2 +b_1  $

$ \tau \dot{x_2} = -\bar{x_2} - \beta x_1 +b_2  $


```python
def eqs(y,t):
    tau=np.array([2,4])
    b=np.array([0.2,0.1])
    beta=0.5
    W=np.array([[0, -beta],[-beta,0]])
    dydt=(-y +  y.dot(W) + b)/tau
    return dydt

y0=np.array([0.1,0.3])
# t = np.linspace(0, 20, 1001)
tmax=0.1
t = np.linspace(0, tmax*1000, tmax*1000+1)

sol = odeint(eqs, y0, t, hmax=1)
plt.style.use('ggplot')
plt.plot(t, sol[:, 0], 'b') #x1
plt.plot(t, sol[:, 1], 'g') #x2
plt.show()
```

Transforming 

$y_1=x_1+x_2$ 
and 
$y_2=x_1-x_2$ 

we get 

$ \tau \dot{y_1} = -\bar{y_1} - \beta y_1 + (b_1+b_2)  $

$ \tau \dot{y_2} = -\bar{y_2} + \beta y_2 + (b_1-b_2)  $

```python

plt.plot(t, sol[:, 0] + sol[:, 1], 'b') #x1+x2
plt.plot(t, sol[:, 0] - sol[:, 1], 'g') #x1-x2
plt.show() 
```

## Network
$ \tau \dot{\bar{r}} = -\bar{r} + W\bar{r} + \bar{i} $

## Misha's paper

```python
def eqs(y,t):
    tau_e=20.0
    tau_i=10.0
    Jee=40.0
    Jei=25.0
    Jie=30.0
    Jii=15.0
    E,I = y
    dEdt = (-E+g_e(Jee *E - Jei * I + e(t)))/tau_e
    dIdt = (-I+g_i(Jie *E - Jii * I + i(t)))/tau_i
    dydt=[dEdt, dIdt]
    return dydt

T=120

def e(t):
    return 0.1
def i(t):
    return 0.1*np.sin(2*np.pi*t/T)
def g_e(x):
    if x>0:
        return np.tanh(x)
    return 0
def g_i(x):
    if x>0:
        return np.tanh(x)
    return 0


y0=np.array([0.,0.])
tmax=1.
t = np.linspace(0, tmax*1000, tmax*1000+1)
sol = odeint(eqs, y0, t)
f, axarr = plt.subplots(2, sharex=True,figsize=(10,10))
axarr[0].plot(t, sol[:, 0], 'b')
axarr[0].plot(t, sol[:, 1], 'g')
axarr[1].plot(t, i(t)*0.1, 'r')
plt.show()
```
